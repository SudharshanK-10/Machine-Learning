```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('iris.csv')
data
```

```python
attributes = list(data.columns)
attributes
```

```python
input = []

for i,attribute in enumerate(attributes):
    if i == len(attributes)-1 :
        break
    input.append(data[attribute].values)

input = np.array(input).T
input
```

```python
output = []
seen = []

for value in data[attributes[-1]].values:
    row = [0,0,0]

    if seen.count(value) > 0:
        row[seen.index(value)] = 1
        output.append(row)
    else:
        seen.append(value)
        row[seen.index(value)] = 1
        output.append(row)

output
```

```python
weights1 = np.random.random((4,5))
bias1 = np.random.random((1,5))

weights2 = np.random.random((5,3))
bias2 = np.random.random((1,3))
```

```python
def sigmoid(x,der=False):
    if der == True:
        return x * (1 - x)
    
    return 1 / (1 + np.exp(-x))
```

```python
predicted = []
learning_rate = 0.001

for epoch in range(10000):

    weighted_sum1 = np.dot(input,weights1) + bias1
    curr_output1 = sigmoid(weighted_sum1)

    weighted_sum2 = np.dot(curr_output1,weights2) + bias2
    curr_output2 = sigmoid(weighted_sum2)

    error2 = output - curr_output2
    adjust2 = error2 * sigmoid(curr_output2,True)

    error1 = np.dot(adjust2,weights2.T)
    adjust1 = error1 * sigmoid(curr_output1,True)

    weights1 += np.dot(input.T,adjust1) * learning_rate
    bias1 += np.sum(adjust1,axis=0) * learning_rate

    weights2 += np.dot(curr_output1.T,adjust2) * learning_rate
    bias2 += np.sum(adjust2,axis=0) * learning_rate
    
    predicted = curr_output2

predicted
```
